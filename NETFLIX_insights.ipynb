# **Cotigorized Vote_Average colomn**

We would cut the Vote_Average value and make 4 catigories : popular average below_avg not_popular to describe it more using categorize_col() function provided above.

```python

def catigorize_col(df,col,labels):

  edge = [df[col].describe()['min'],
  df[col].describe()['25%'],
  df[col].describe()['50%'],
  df[col].describe()['75%'],
  df[col].describe()['max']]

  df[col]=pd.cut(df[col],bins=edge,labels=labels,duplicates ='drop')
  return df
```

```python
labels = ['not_popular','below_avg','average','popular']
catigorize_col(df,'Vote_Average',labels)

df['Vote_Average'].unique()
```
